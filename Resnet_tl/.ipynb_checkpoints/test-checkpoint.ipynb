```python
#MAE

temp=sio.loadmat('../data/weight_matrix.mat')
weight_matrix=temp['weight_matrix']

test_total_ae=0
test_total_count=0
start=time.time()
for i in range(401,500+1):
    im = scipy.misc.imread('../data/origin_images/%.4d.png' % (i))
    im = np.float32(im)
    batch_images=[im]         
    test_output_eval= sess.run(test_output,feed_dict={test_input: batch_images}) 
    im_out=test_output_eval[0]
    h,w,channel=im_out.shape
    im_out=(np.float64(im_out)+1)/2*255
    im_out[im_out>255]=255
    im_out[im_out<0]=0
#     scipy.misc.imsave('model_result_ResNet/%.4d.png'%(i),im_out/255)
    im_groundtruth1 = scipy.misc.imread('../data/gt_images/%.4d_1.png' % (i))
    im_groundtruth1 = np.float64(im_groundtruth1)
    im_groundtruth2 = scipy.misc.imread('../data/gt_images/%.4d_2.png' % (i))
    im_groundtruth2 = np.float64(im_groundtruth2)
    im_groundtruth3 = scipy.misc.imread('../data/gt_images/%.4d_3.png' % (i))
    im_groundtruth3 = np.float64(im_groundtruth3)
    im_groundtruth4 = scipy.misc.imread('../data/gt_images/%.4d_4.png' % (i))
    im_groundtruth4 = np.float64(im_groundtruth4)
    im_groundtruth5 = scipy.misc.imread('../data/gt_images/%.4d_5.png' % (i))
    im_groundtruth5 = np.float64(im_groundtruth5)
    image_ae = np.sum(np.abs(im_out-im_groundtruth1))*weight_matrix[i-1,0] \
                +np.sum(np.abs(im_out-im_groundtruth2))*weight_matrix[i-1,1] \
                +np.sum(np.abs(im_out-im_groundtruth3))*weight_matrix[i-1,2] \
                +np.sum(np.abs(im_out-im_groundtruth4))*weight_matrix[i-1,3] \
                +np.sum(np.abs(im_out-im_groundtruth5))*weight_matrix[i-1,4] 
    test_total_ae=test_total_ae + image_ae
    test_total_count=test_total_count+h*w*channel
    
    ### save output
mae=test_total_ae/test_total_count
print 'test MAE: %.3f' % (mae)
print (time.time()-start)/100
```

```python

## RMSE 1

temp=sio.loadmat('../data/weight_matrix.mat')
weight_matrix=temp['weight_matrix']

test_total_se=0
test_total_count=0
for i in range(401,500+1):
    im = scipy.misc.imread('../data/origin_images/%.4d.png' % (i))
    im = np.float32(im)
    batch_images=[im]         
    test_output_eval= sess.run(test_output,feed_dict={test_input: batch_images}) 
    im_out=test_output_eval[0]
    h,w,channel=im_out.shape
    im_out=(np.float64(im_out)+1)/2*255
    im_out[im_out>255]=255
    im_out[im_out<0]=0
    #scipy.misc.imsave('model_result/%.4d.png'%(i),im_out/255)
    im_groundtruth1 = scipy.misc.imread('../data/gt_images/%.4d_1.png' % (i))
    im_groundtruth1 = np.float64(im_groundtruth1)
    im_groundtruth2 = scipy.misc.imread('../data/gt_images/%.4d_2.png' % (i))
    im_groundtruth2 = np.float64(im_groundtruth2)
    im_groundtruth3 = scipy.misc.imread('../data/gt_images/%.4d_3.png' % (i))
    im_groundtruth3 = np.float64(im_groundtruth3)
    im_groundtruth4 = scipy.misc.imread('../data/gt_images/%.4d_4.png' % (i))
    im_groundtruth4 = np.float64(im_groundtruth4)
    im_groundtruth5 = scipy.misc.imread('../data/gt_images/%.4d_5.png' % (i))
    im_groundtruth5 = np.float64(im_groundtruth5)
    image_se = np.sum(np.square(im_out-im_groundtruth1))*weight_matrix[i-1,0] \
                +np.sum(np.square(im_out-im_groundtruth2))*weight_matrix[i-1,1] \
                +np.sum(np.square(im_out-im_groundtruth3))*weight_matrix[i-1,2] \
                +np.sum(np.square(im_out-im_groundtruth4))*weight_matrix[i-1,3] \
                +np.sum(np.square(im_out-im_groundtruth5))*weight_matrix[i-1,4] 
    test_total_se=test_total_se + image_se
    test_total_count=test_total_count+h*w*channel
    
print 'test RMSE: %.3f' % (np.sqrt(test_total_se/test_total_count))
```

```python
total_psnr1=0.0
total_ssim=0.0
def get_start_points(l):
    patch=96
    overlap=30
    start_points=list()
    start_points.append(0)
    while 1:
        next_point=start_points[len(start_points)-1]+patch-overlap
        if l-next_point<=patch:
            next_point=l-patch
            start_points.append(next_point)
            break
        else:
            start_points.append(next_point)
    return start_points
        
for i in range(401,410+1):
    im = scipy.misc.imread('../data/origin_images/%.4d.png' % (i))
    im = np.float32(im)
    
    h,w=im.shape[0:2]
    row_points=get_start_points(h)
    col_points=get_start_points(w)
    count=np.zeros((h,w,3))
    im_total1=np.zeros((h,w,3))
    for row in row_points:
        for col in col_points:
            count[row:row+128,col:col+128,:]=count[row:row+128,col:col+128,:]+1
            batch_images=[im[row:row+128,col:col+128,:]]  
            t1_eval= sess.run(test_output,feed_dict={test_input: batch_images}) 
            im_out1=t1_eval[0]
            im_out1=np.float64(im_out1)/255
            im_out1[im_out1>1]=1
            im_out1[im_out1<0]=0
            im_total1[row:row+128,col:col+128,:]=im_total1[row:row+128,col:col+128,:]+im_out1
                 
    im_out1=im_total1/count
    im_groundtruth = scipy.misc.imread('../data/gt_images/%.4d_1.png' % (i))
    im_groundtruth = np.float64(im_groundtruth)/255
    total_psnr1=total_psnr1+psnr(im_out1,im_groundtruth)

#     if i==1:
#         path='%s/%.7d.png' % ('test',1)
#         im_out=np.concatenate((im_rgb,im_groundtruth,im_out1,im_out2),axis=1)
#         plt.figure(figsize=(100, 25))
#         plt.imshow(im_out)
#         #plt.imshow(im_out)
#         scipy.misc.imsave(path,im_out)
print 'average PSNR1: %.3f' % (total_psnr1/10)
```
